```python
le=LabelEncoder()

df['diagnosis']=le.fit_transform(df['diagnosis'].astype(str))

df.head()
```

```python
x=df.drop(columns=['diagnosis'])
y=df['diagnosis']
```

```python
print(df.columns)
```

```python
from sklearn.preprocessing import StandardScaler
scaler =StandardScaler()
x=scaler.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30)
```

```python
from sklearn.tree import DecisionTreeClassifier as DTC
from sklearn import tree

modeldtc=DTC(criterion='entropy',min_samples_split=2)
preddtc=modeldtc.fit(x_train,y_train).predict(x_test)
acc=np.mean(preddtc==y_test)
acc
```

```python
tree.plot_tree(modeldtc)
```
